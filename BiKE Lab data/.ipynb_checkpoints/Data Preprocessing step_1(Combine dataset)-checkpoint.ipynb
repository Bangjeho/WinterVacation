```python
import pandas as pd
import numpy as np
```

**Load Dataset**

```python
Reactome_id = pd.read_excel('03_Reactome entrez genes ids.xlsx', header = None)
Kegg_id = pd.read_excel('05_KEGG entrez genes ids.xlsx', header = None)
Biocarta_id = pd.read_excel('07_BioCarta entrez genes ids.xlsx', header = None)
```

**Checking Dataset infomation**

```python
Kegg_id
```

```python
Biocarta_id.info()
```

**Combine Dataset**

```python
Reactome_N = Reactome_id.drop([0,1],axis = 1)
Kegg_N = Kegg_id.drop([0,1],axis = 1)
Biocarta_N = Biocarta_id.drop([0,1], axis = 1)

ReacName = Reactome_id[[0,1]]
KeggName = Kegg_id[[0,1]]
BioName = Biocarta_id[[0,1]]
```

```python
Kegg_N
```

```python
concat_df_kegg = pd.DataFrame(columns= np.arange(Kegg_N.columns[-1]+1,Reactome_id.columns.size))
concat_df_bio = pd.DataFrame(columns= np.arange(Biocarta_N.columns[-1]+1,Reactome_id.columns.size))
```

```python
Kegg_N = pd.concat([Kegg_N,concat_df_kegg],axis = 1)
Biocarta_N = pd.concat([Biocarta_N,concat_df_bio],axis = 1)
```

```python
Kegg_N
```

```python
overlap_index = []
for i in range(len(Reactome_N)):
    for j in range(len(Kegg_N)):
        if Reactome_N.iloc[[i]].equals(Kegg_N.iloc[[j]]) :
            overlap_index.append(j)
```

```python
overlap_index
```

```python
all_dataName = pd.concat([ReacName,KeggName],axis = 0)
all_Data_N = pd.concat([Reactome_N,Kegg_N],axis = 0)

all_dataName.reset_index(drop = True)
all_Data_N.reset_index(drop = True)
```

```python
overlap_index = []
for i in range(len(all_Data_N)):
    for j in range(len(Biocarta_N)):
        if all_Data_N.iloc[[i]].equals(Biocarta_N.iloc[[j]]) :
            overlap_index.append(j)

print(overlap_index)
```

```python
all_Data_N = pd.concat([all_Data_N,Biocarta_N],axis = 0)
all_dataName = pd.concat([all_dataName,BioName],axis = 0)

all_dataName.reset_index(drop = True)
all_Data_N.reset_index(drop = True)
```

```python
all_Data = pd.concat([all_dataName,all_Data_N],axis = 1)

all_Data.reset_index(drop = True)
```

```python
Reactome_Sym = pd.read_excel("02_Reactome gene symbols.xlsx",header = None)
Kegg_Sym = pd.read_excel("04_KEGG gene sets and symbols.xlsx",header = None)
Biocarta_Sym = pd.read_excel("06_BioCarta gene symbols.xlsx",header = None)
```

```python
all_data_Sym = pd.concat([Reactome_Sym,Kegg_Sym,Biocarta_Sym],axis = 0)
```

```python
all_data_Sym.reset_index(drop = True)
```

**from DataFrame to excel file**

```python
all_data_Sym.to_excel("09_AllDB gene symbols.xlsx", index = None, header = None)
all_Data.to_excel("10_AllDB entrez genes ids.xlsx", index = None, header = None)
```
